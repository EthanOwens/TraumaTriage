```python
import statistics
import math
import csv


class Statistics:
    def __init__(self, path, prefix):
        self.path = path
        self.prefix = prefix
        
        
    def get_stats(self, spec): #grab data, I don't understand what is being passed here. What am I getting?
        #recieving list of classifiers and 2D list of CMs

        #find how many runs by variaitons of model in data
        k = len(spec) #number of model types
        j = len(spec[0]) #runs
        
        #create necessary lists
        mean = [[_ for x in range(j)] for y in range(k)] #Mean for each classifier type not necessarily run
        var = [[_ for x in range(j)] for y in range(k)] #Variance for each classifier type and run
        std_dev = [[_ for x in range(j)] for y in range(k)] #Standard Deviation for each classifier type and run

        for a in range(k):
            for b in range(j):
                #number of observations
                n = len(spec[a])

                #mean of data
                mean[a][b] = sum(spec[a]) / n 

                #squared deviations
                deviations = [(x - mean[a][b])**2 for x in spec[a]]

                #Variance
                var[a][b] = sum(deviations) / n

                #Standard deviation
                std_dev[a][b] = math.sqrt(var[a][b])


                #P-values implementation saved for later

        #save statistics to files (for python 3)
        with open(f'{self.path}\\{self.prefix}-Mean.csv', "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerows(mean)

        with open(f'{self.path}\\{self.prefix}-Variance.csv', "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerows(var)

        with open(f'{self.path}\\{self.prefix}-Std_Dev.csv', "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerows(std_dev)
```
