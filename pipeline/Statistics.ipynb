```python
import statistics
import math
import csv
import pandas as pd


class Statistics:
    def __init__(self, path, prefix):
        self.path = path
        self.prefix = prefix
        
    def set_prefix(self, prefix):
        self.prefix = prefix
        
        
    def get_stats(self, spec, params): #grab data, I don't understand what is being passed here. What am I getting?
        #recieving list of classifiers and 2D list of CMs

        #find how many runs by variaitons of model in data
        k = len(spec) #number of model types
        j = len(spec[0]) #runs
        
        #create necessary lists
        mean = [[_ for x in range(j)] for y in range(k)] #Mean for each classifier type not necessarily run
        var = [[_ for x in range(j)] for y in range(k)] #Variance for each classifier type and run
        std_dev = [[_ for x in range(j)] for y in range(k)] #Standard Deviation for each classifier type and run

        
        NUM_SIG_STATS = 3 #the number of stats we are documenting. Will increase if we need to add more later
        stats_avg = [[_ for x in range(NUM_SIG_STATS + 1)] for y in range(k)] #creates a 4 length list for each clf
                                                                              
        
        for a in range(k):
            for b in range(j):
                #number of observations
                n = len(spec[a])

                #mean of data
                mean[a][b] = sum(spec[a]) / n 

                #squared deviations
                deviations = [(x - mean[a][b])**2 for x in spec[a]]

                #Variance
                var[a][b] = sum(deviations) / n

                #Standard deviation
                std_dev[a][b] = math.sqrt(var[a][b])
                
                #P-values implementation saved for later
            
            
            #[parameters, mean, variance, standard deviation]
            pair = params[a]
            stats_avg[a] = [pair, sum(mean[a])/j, sum(var[a])/j, sum(std_dev[a])/j]


        #save statistics to files (for python 3)
        self.filepath = f'{self.path}\\{self.prefix}-statistics.csv'
        with open(self.filepath, "w", newline="") as f:
            header = ['parameters', 'specificty mean', 'specificty variance', 'specificty standard dev']
            writer = csv.writer(f)
            
            writer.writerow(header)
            writer.writerows(stats_avg)
            
        return mean, var, std_dev
            
    def get_last(self):
        try:
            data = pd.read_csv(self.filepath)
            return data
        except:
            return "No filepath on stack"
            
            
    def import_stats(self, file):
        data = pd.read_csv(f'{self.path}/{file}')
        print(data)
        
```
